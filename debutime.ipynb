```python
args = dict()
args['command'] = 'with_fairness'
args['df_name'] = 'vital'
args['module'] = 'gru'
args['hidden_dim'] = 24
args['num_layer'] = 3
args['iterations'] = 1000
args['batch_size'] = 10
args['num_epochs'] = 10
args['seq_len'] = 12
args['S'] = 'Gender'
args['Y'] = 'Mortality'
args['underprivileged_value'] = 'Female'
args['desirable_value'] = 'No'
args['lamda_val'] = 0.5
df, ohe, scaler, discrete_columns_ordereddict, continuous_columns_list, ori_data, S_start_index, Y_start_index, underpriv_index, priv_index, undesire_index, desire_index = real_data_loading(args)
print(args['df_name'] + ' dataset is ready.')
```

```python
args['S_start_index'] = S_start_index
args['Y_start_index'] = Y_start_index
args['underpriv_index'] = underpriv_index
args['priv_index'] = priv_index
args['undesire_index'] = undesire_index
args['desire_index'] = desire_index
generated_data = timegan(ori_data, args)   
print('Finish Synthetic Data Generation')
synthetic_data_reverse = []
for row in generated_data:
    for batch in row:
        synthetic_data_reverse.append(list(batch))
synthetic_data = synthetic_data_reverse[::-1]
```
